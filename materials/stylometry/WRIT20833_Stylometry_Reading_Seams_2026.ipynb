## The corpus
Three short texts: one we wrote to sound human, one typical AI-assistant prose, and the human-edited AI email from the PDF. Later, swap in your **own** samples (Part C).

```python
# Our small corpus (edit these later with your own texts)
human = (
    "I keep a cracked mug on my desk. It leaks a little. I keep it anyway because my "
    "sister made it in a class she hated and finished anyway. That stubbornness is the point."
)

ai = (
    "That is a great question! There are several factors to consider. First, it is important "
    "to weigh the pros and cons. Additionally, you may want to explore a few options. "
    "Ultimately, the best choice depends on your specific needs and goals."
)

hybrid = (
    "Your line about AI writing being standardized and averaged is one of the stronger "
    "moments in the piece, but my editor flagged that readers may want a concrete picture "
    "of what that actually looks like on the page. Could you share a quick example or two? "
    "Anything that comes to mind: a turn of phrase AI keeps reaching for, a structural tic, "
    "a passage you have encountered in student work that gave it away."
)

texts = {"human": human, "ai": ai, "hybrid": hybrid}
for name, t in texts.items():
    print(name, "->", len(t.split()), "words")
```

## Feature 1 — Function-word share *(the HW1 inversion)*
In HW1, stopwords were **noise to remove**. In stylometry they are the **signal**: how often a writer leans on tiny connective words is a fingerprint of style.

```python
# Function words = the little connective words. (Same list HW1 threw away.)
function_words = ["the","a","an","and","but","or","of","to","in","on","for","with",
                  "is","are","was","it","that","this","you","your","may","might",
                  "several","additionally","ultimately","first","also","whatever"]

def function_word_share(text):
    words = text.lower().split()
    hits = [w.strip(".,!?;:") for w in words]
    n_func = sum(1 for w in hits if w in function_words)
    return round(n_func / len(words), 3)

for name, t in texts.items():
    print(name, "function-word share:", function_word_share(t))
```

## Feature 2 — Type-token ratio (lexical diversity)
Distinct words ÷ total words. Lower can mean more repetition / more 'averaged' phrasing.

```python
def type_token_ratio(text):
    words = [w.strip(".,!?;:").lower() for w in text.split()]
    return round(len(set(words)) / len(words), 3)

for name, t in texts.items():
    print(name, "type-token ratio:", type_token_ratio(t))
```

## Feature 3 — Average sentence length
Split on sentence-ending punctuation and average the word counts. Very even, mid-length sentences are a common AI tell.

```python
def avg_sentence_length(text):
    import re
    sentences = [s for s in re.split(r"[.!?]", text) if s.strip()]
    lengths = [len(s.split()) for s in sentences]
    return round(sum(lengths) / len(lengths), 1)

for name, t in texts.items():
    print(name, "avg words/sentence:", avg_sentence_length(t))
```

## Feature 4 — Hedging / softening phrases
The 'burden-reducing' language the PDF flagged: count how often each text reaches for it.

```python
hedges = ["a great question","several factors","it is important","you may want",
          "ultimately","a quick example","anything that comes to mind","whatever feels",
          "perhaps","might want"]

def hedging_count(text):
    t = text.lower()
    return sum(t.count(h) for h in hedges)

for name, t in texts.items():
    print(name, "hedging hits:", hedging_count(t))
```

## Put the signature together

```python
header = "text".ljust(8) + "func".rjust(7) + "ttr".rjust(7) + "sent".rjust(7) + "hedge".rjust(7)
print(header)
for name, t in texts.items():
    row = (name.ljust(8) + str(function_word_share(t)).rjust(7)
           + str(type_token_ratio(t)).rjust(7) + str(avg_sentence_length(t)).rjust(7)
           + str(hedging_count(t)).rjust(7))
    print(row)
```

### Optional: a quick bar chart

```python
import matplotlib.pyplot as plt
names = list(texts)
values = [hedging_count(texts[n]) for n in names]
plt.bar(names, values)
plt.title("Hedging / softening hits by text")
plt.ylabel("count")
plt.show()
```
